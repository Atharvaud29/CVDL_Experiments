```python
# Importing necessary libraries
from ultralytics import YOLO  # import the YOLO model class from the ultralytics library to perform object detection
from PIL import Image  # used to open and work with image files in Python
import requests  # allows us to fetch data from the internet, like downloading an image using its URL
from io import BytesIO  # helps convert the downloaded image bytes into a format that PIL can open
import matplotlib.pyplot as plt  # used to display images and plots inside the notebook
```

```python
# Load the YOLOv8 model
model = YOLO('yolov8n.pt')  # this line processes image or data
```

```python
# Testing model
image_url = 'https://ultralytics.com/images/zidane.jpg' # URL for image is provided here
response = requests.get(image_url) # send a request to the URL and download the image
img = Image.open(BytesIO(response.content)) # convert the downloaded image bytes into a PIL image
results = model(img) # pass the image to the YOLO model to detect objects
plt.imshow(results[0].plot()) # display the image with detected objects (boxes, labels)
plt.axis('off') # hide axis markings for cleaner output
plt.show() # show the final plotted image
```

```python
from google.colab import files # import module to upload files from local system
uploaded = files.upload() # open a file upload dialog and store uploaded files in a dictionary
img_path = next(iter(uploaded)) # get the name of the first uploaded file
img = Image.open(img_path) # open the uploaded image using PIL
# Perform object detection
results = model(img) # run the YOLO model on the uploaded image
# Display the image with bounding boxes
plt.imshow(results[0].plot()) # show the image with detected object boxes and labels
plt.axis('off') # remove axis ticks and labels for a cleaner look
plt.show() # display the final image
```
