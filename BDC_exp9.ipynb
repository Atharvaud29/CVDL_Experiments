```r
# 1. Load the In-Built Dataset
data(mtcars)
head(mtcars)
```

```r
# 2. Visualizations Using Base R Functions
# Histogram: Distribution of Miles per Gallon (mpg)
hist(mtcars$mpg,
     main = "Histogram of MPG (Base R)",
     xlab = "Miles per Gallon",
     col = "lightblue", border = "black")
```

```r
# Boxplot: Miles per Gallon (mpg) by Number of Cylinders (cyl)
boxplot(mpg ~ as.factor(cyl), data = mtcars,
        main = "Boxplot of MPG by Cylinders (Base R)",
        xlab = "Number of Cylinders",
        ylab = "Miles per Gallon",
        col = c("orange", "green", "blue"))
```

```r
# 3. Visualizations Using ggplot2
library(ggplot2)
```

```r
# Density Plot with ggplot2: Distribution of MPG by Number of Cylinders
density_plot <- ggplot(mtcars, aes(x = mpg, fill = as.factor(cyl))) +
  geom_density(alpha = 0.5) +
  labs(title = "Density Plot of MPG by Cylinders (ggplot2)",
       x = "Miles per Gallon", y = "Density",
       fill = "Cylinders") +
  theme_minimal()
print(density_plot)
```

extra code

```r
# 1) Bar chart: mean MPG by cylinder count
mean_mpg <- aggregate(mpg ~ cyl, data = mtcars, FUN = mean)
barplot(mean_mpg$mpg,
        names.arg = mean_mpg$cyl,
        col = "steelblue",
        main = "Mean MPG by Cylinder Count",
        xlab = "Cylinders",
        ylab = "Mean MPG")
```

```r
# 2) Scatter plot + regression: MPG vs Weight
plot(mtcars$wt, mtcars$mpg,
     pch = 19, col = "darkgreen",
     main = "MPG vs Weight",
     xlab = "Weight (1000 lbs)",
     ylab = "MPG")
model <- lm(mpg ~ wt, data = mtcars)
abline(model, lty = 2, lwd = 2, col = "red")
r2 <- summary(model)$r.squared
legend("topright", legend = paste0("R² = ", round(r2, 2)),
       bty = "n")
```

```r
# 3) QQ-plot: check normality of MPG
qqnorm(mtcars$mpg,
       main = "QQ-Plot of MPG")
qqline(mtcars$mpg,
       col = "blue", lwd = 2)
```

```r
# 5) Violin plot: MPG distributions by cylinder
library(ggplot2)
ggplot(mtcars, aes(x = cyl, y = mpg, fill = cyl)) +
  geom_violin(trim = FALSE, show.legend = FALSE) +
  stat_summary(fun = mean, geom = "point", shape = 23, size = 3, fill = "white") +
  labs(title = "Violin Plot of MPG by Cylinders",
       x = "Cylinders",
       y = "Miles per Gallon") +
  theme_minimal()
```
