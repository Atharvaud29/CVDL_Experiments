```r
library(ggplot2)
library(dplyr)
library(readr)
```

```r
tips <- read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
```

```r
head(tips)
```

```r
summary(tips)
```

```r
# Bar Plot: Average Tip by Day
tips %>%
  group_by(day) %>%
  summarise(avg_tip = mean(tip)) %>%
  ggplot(aes(x=reorder(day, avg_tip), y=avg_tip, fill=day)) +
  geom_bar(stat="identity", width=0.6) +
  labs(title="Average Tip Given by Day", x="Day", y="Average Tip ($)", fill="Day") +
  theme_minimal() +
  theme(plot.title = element_text(hjust = 0.5, size=16, face="bold")) +
  scale_fill_brewer(palette="Set2")
```

```r
# Scatter Plot: Total Bill vs Tip
ggplot(tips, aes(x=total_bill, y=tip, color=sex)) +
  geom_point(size=4, alpha=0.7) +
  labs(title="Total Bill vs. Tip Amount", x="Total Bill ($)", y="Tip Amount ($)", color="Gender") +
  theme_minimal() +
  theme(plot.title = element_text(hjust = 0.5, size=16, face="bold"))
```

```r
# Box Plot: Tip Percentage by Gender
ggplot(tips, aes(x=sex, y=tip/total_bill * 100, fill=sex)) +
  geom_boxplot() +
  labs(title="Tip Percentage by Gender", x="Gender", y="Tip Percentage (%)", fill="Gender") +
  theme_minimal() +
  theme(plot.title = element_text(hjust = 0.5, size=16, face="bold")) +
  scale_fill_brewer(palette="Dark2")
```

```r
# Density Plot: Distribution of Total Bill
ggplot(tips, aes(x=total_bill, fill=day)) +
  geom_density(alpha=0.6) +
  labs(title="Distribution of Total Bill", x="Total Bill ($)", fill="Day") +
  theme_minimal() +
  theme(plot.title = element_text(hjust = 0.5, size=16, face="bold")) +
  scale_fill_brewer(palette="Pastel1")
```
